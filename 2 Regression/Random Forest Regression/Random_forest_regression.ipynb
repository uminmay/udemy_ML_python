# Random Forest Regression

## Ensemble Learning

### Importing Libraries and Dataset

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

### Training Random Forest on entire dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state=0) ## number of trees
regressor.fit(X, y)
```

### Predicting results from our model

```python
regressor.predict([[6.5]])
```

### Visualising the model HIres

```python
X_grid = np.arange(min(X), max(X),0.1)
X_grid = X_grid.reshape((len(X_grid)),1)
plt.scatter(X, y, color='red')
plt.plot(X_grid,regressor.predict(X_grid),color='blue')
plt.title('Bluff/Truth..Random Forest')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

### Visualisation LOWres

```python
plt.scatter(X, y, color='red')
plt.plot(X,regressor.predict(X),color='blue')
plt.title('Bluff/Truth..Random Forest')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
