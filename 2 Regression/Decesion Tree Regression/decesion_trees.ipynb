# Decesion Tree Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Position_Salaries.csv')
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

## Training Decesion tree model on whole dataset

```python
###NO FEATURE SCALIN G HERE
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0) ##parameterTunnings
regressor.fit(X,y)
```

## Predicting a new result

```python
regressor.predict([[6.5]])
```

## Visualising the decesion tree results(HiRes)

```python
X_grid = np.arange(min(X), max(X),0.1)
X_grid = X_grid.reshape((len(X_grid)),1)
plt.scatter(X, y, color='red')
plt.plot(X_grid,regressor.predict(X_grid),color='blue')
plt.title('Bluff/Truth..Decesipm')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## Visualising the decesion tree results(LOWRES)

```python
plt.scatter(X, y, color='red')
plt.plot(X,regressor.predict(X),color='blue')
plt.title('Bluff/Truth..Decesipm')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
